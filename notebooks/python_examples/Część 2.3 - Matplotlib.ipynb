# Matplotlib - pakiet wizualizacji danych

Matplotlib jest podstawowym narzędziem do tworzenia wszelakich wizualizacji w Pythonie. Jest również w całości wspierany przez Jupyter, jako warstwie prezentacji dla wykresów.

## Podstawowe wykresy matplotlib

Teraz pokażemy jak w prosty sposób użyć pakietu matplotlib do tworzenia prostych wykresów.
Dodatkową komendą potrzebną do użycia matplotlibowych obiektów w Jupyter jest tzw. słowo magiczne %matplotlib inline

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = [6, 8, 10, 14, 18]
y = [7, 9, 13, 17.5, 18]

plt.figure()
plt.title('Wykres przykładowy')
plt.xlabel('Coś x')
plt.ylabel('Coś y')
plt.plot(x, y, 'k.')
plt.axis([0,25,0,25])
plt.grid(True)
plt.show()
```

Wykresy potrafią również używać obiektów numpy i funkcji z tego pakietu


```python

x = np.linspace(0,7*np.pi, num=100)
y = np.sin(x)

plt.figure()
plt.title(r'Wykres $\sin(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y, 'r-')
plt.axis([0,25,-1,1])
plt.grid(True)
plt.show()
```

## Wykresy wielokrotne

Możliwe jest prezentowanie wielu wykresów na raz. Wtedy przydaje się rozróżnić je za pomocą odpowiednich symboli

* '-' wykres ciągłą linią
* '--' wykres linią przerywaną
* '-.' wykres linią przerywana kropkami
* ':' wykres kropkowy
* '.' wykres - punkty oznaczone kropką
* ',' wykres - punkty oznaczone pikselem
* 's' wykres - punkty oznaczone kwadratem
* 'o' wykres - punkty oznaczone kołem
* '*' wykres - punkty oznaczone gwiazdką
* 'p' wykres - punkty oznaczone pięciokątem
* 'h', 'H' wykres - punkty oznaczone sześciokątem
* 'd', 'D' wykres - punkty oznaczone diamentem
* 'x' wykres - punkty oznaczone x
* '|', '_' wykres - punkty oznaczone linią pionową, poziomą
* 'v', '^', '<' , '>' wykres - punkty oznaczone trójkątami
* '1', '2', '3', '4' wykres - punkty oznaczone trójkątnymi krzyżami

Oraz kolory

* 'b' - niebieskie
* 'g' - zielony
* 'r' - czerwony
* 'c' - cyjan
* 'm' - magento
* 'y' - żółty
* 'k' - czarny
* 'w' - biały

```python
x = np.linspace(0, 10)

plt.plot(x, np.sin(x), '1g', x, np.cos(x), ',r', x, -np.sin(x), 'db')
plt.show()
```

## Wykresy dzielące okno

Można również tworzyć wykresy będące połączeniem kilku wykresów obok siebie

wykorzystywana do tego jest komenda subplot której przekazywane są 3 cyfry zapisane jak jedna liczba


```python
x = np.linspace(0, 10)

plt.figure(1)
plt.subplot(221)
plt.plot(x, np.sin(x), '1g')
plt.subplot(222)
plt.plot(x, np.cos(x), ',r')
plt.subplot(223)
plt.plot(x, -np.sin(x), 'db')
plt.show()
```

## Wykresy 3D

Nieco trudnijszym zadaniem są wykresy 3D. 

Tak jak do wytworzenia wykresu 2D potrzebne są dwa wektory sparowanych wartości x oraz y, tak dla dowolnego wykresu 3D dane powinny być zgromadzone w (tym razem) 3 kolekcjach, z których każda jest macierzą. Najpierw więc należy transformować wektory jednowymiarowe na macierze argumentów

Tak jak do wytworzenia wykresu 2D potrzebne są dwa wektory sparowanych wartości x oraz y, tak dla dowolnego wykresu 3D dane powinny być zgromadzone w (tym razem) 3 kolekcjach, z których każda jest macierzą. Najpierw więc należy transformować wektory jednowymiarowe na macierze argumentów

```python
X, Y = np.meshgrid(x, y)
print('X\n', X)
print('Y\n', Y)
```

Jak widać tak utworzone macierze są kopiami x, y odpowiednio wierszowo oraz kolumnowo. Za jej pomocą utwórzmy macierz wartości

```python
Z = X*X+Y*Y
print('Z\n',Z)
```

Utwórzmy teraz powierzchniowy

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm as cm
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
plt.show()
```

W ten sposób można oddawać różne finezyjne kształty geometrii trójwymiarowej oraz stosować podstawowe tranformacje matematyczne

```python
alpha = np.linspace(0,2*np.pi,num=100)
beta = np.linspace(-np.pi,np.pi,num=150)

alpha, beta = np.meshgrid(alpha,beta)

x = np.cos(alpha)*np.cos(beta)
y = np.cos(alpha)*np.sin(beta)
z = np.sin(alpha)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x,y,z, rstride=1,cstride=1,cmap=cm.jet)
plt.show()
```
